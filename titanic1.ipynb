```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import warnings
from seaborn import load_dataset
warnings.filterwarnings('ignore')

titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic.info()
```

```python
titanic.drop(['deck'],axis=1,inplace=True)
titanic.shape
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='survived', data=titanic, palette='Set1')
plt.title("Survived and Not Survived")
plt.xlabel("Survival")
plt.ylabel("Count")
```

```python
print("Average Age : ",round(titanic['age'].mean(),2))
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(titanic['age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python

# Plot survival based on age with adjusted x-ticks
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic, x='age', hue='survived', bins=30, kde=True, palette='Set1')
plt.title('Survival Based on Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.xticks(range(0, 90, 5))  # Set x-ticks at intervals of 5 years
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
# Class Distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='class', data=titanic, palette='Set2')
plt.title('Class Distribution')
plt.show()
```

```python
# Convert 'survived' column to string data type
titanic['survived'] = titanic['survived'].astype(str)

# Plot class-wise survival
plt.figure(figsize=(8, 5))
sns.countplot(x='pclass', hue='survived', data=titanic, palette='Set1')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
print("Average Fare : ",round(titanic['fare'].mean(),2))
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(titanic['fare'], bins=30, kde=True, color='skyblue')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(x='fare', data=titanic, hue='survived', bins=30, kde=True, palette='Set1')
plt.title('Survival Based on Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
